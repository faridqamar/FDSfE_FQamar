```python
import pandas as pd
import geopandas as gpd
```

```python
arrests = pd.read_csv("https://github.com/phillydao/phillydao-public-data/blob/main/docs/data/arrest_data_daily_by_zip.csv?raw=true")
arrests
```

```python
arrests.info()
```

```python
arrests.head()
```

```python
arrests.tail()
```

```python
ziparrests = arrests.groupby("ZCTA_GEOID").sum()
```

```python
zips = gpd.GeoDataFrame.from_file("https://opendata.arcgis.com/api/v3/datasets/ad233733f89242af9a1f9e52d60d1e4e_0/downloads/data?format=shp&spatialRefId=4326")

```

```python
zips.head()
```

```python
zips.crs
```

```python
ax = zips.plot()
ax.set_xlabel("longitude (deg)")
ax.set_ylabel("latitude (deg)");
```

```python
phillyzips = gpd.GeoDataFrame.from_file("https://opendata.arcgis.com/api/v3/datasets/b54ec5210cee41c3a884c9086f7af1be_0/downloads/data?format=shp&spatialRefId=4326")
phillyzips.head()
```

```python
ax = phillyzips.plot()

ax.set_xlabel("longitude (deg)")
ax.set_ylabel("latitude (deg)");
```

```python
ziparrestsgpd = phillyzips.merge(ziparrests, left_on="CODE", right_on="ZCTA_GEOID")
```

```python
ziparrestsgpd.info()
```

```python
ziparrestsgpd.plot?
```

```python
ax = ziparrestsgpd.plot(column="Auto Theft", legend=True);
ax.axis('off');
```

```python
ax = ziparrestsgpd.plot(column="Illegal Dumping/Littering", legend=True);
ax.axis('off');
```

```python
ctpop = gpd.GeoDataFrame.from_file("https://opendata.arcgis.com/api/v3/datasets/20332a074f0446b3b3190ba9d68b863e_0/downloads/data?format=geojson&spatialRefId=4326")
```

```python
ctpop.info()
```

```python
ctpop.head()
```

```python
ax = ctpop.plot(column="COUNT_ALL_RACES_ETHNICITIES")
phillyzips.plot(ax=ax, fc="None", ec="k")
ax.axis("off");
```

```python
ctpop.crs
```

```python
import tobler
```

```python
results = tobler.area_weighted.area_interpolate(source_df=ctpop, target_df=ziparrestsgpd, 
                           intensive_variables=['COUNT_ALL_RACES_ETHNICITIES'])
results
```

```python
ziparrestspop = results.sjoin(ziparrestsgpd)
```

```python
import pylab as plt
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(131)
ax = ziparrestspop.plot(column="COUNT_ALL_RACES_ETHNICITIES", ax=ax)
ax.set_title("population")
ax.axis('off')

ax = fig.add_subplot(132)

ax = ziparrestspop.plot(column="Auto Theft", ax=ax)
ax.set_title("Auto Theft")
ax.axis('off')

ax = fig.add_subplot(133)

ax = ziparrestspop.plot(column="Illegal Dumping/Littering", ax=ax)
ax.set_title("Illegal Dumping/Littering")
ax.axis('off');
```
